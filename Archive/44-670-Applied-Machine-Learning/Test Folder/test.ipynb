```python
import pandas as pd
import pathlib
import os
from pathlib import Path
import matplotlib.pyplot as plt

path = str(pathlib.Path().resolve())
key = '0-60 MPH Time (seconds)'
key2 = 'Horsepower'
key3 = 'Price (in USD)'

for path in Path(path).rglob('*.csv'):
    file = path.name
    path = str(pathlib.Path().resolve())
    filename = Path(file)
    filename_wo_ext = filename.with_suffix('')
    file = str(filename_wo_ext)

df = pd.read_csv(f'{file}.csv')

print(df.keys())
print(df.describe())
```

```python
print(df[key].describe())
df.sort_values(key, inplace=True)

df[key].hist()
plt.xticks(rotation=90)
plt.ylabel('Cost')
plt.xlabel('Seconds')
plt.show()
```

```python
df.groupby(key)[key3].value_counts().unstack().plot.bar(ylabel='Athlete per year')
```
